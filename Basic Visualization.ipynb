```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt('data/data.txt', delimiter='\t')
```

```python
plt.hist(data[:, 2], bins=5, range=(1,6))
plt.title('All Ratings')
plt.show()
```

```python
def get_ratings(movie_ids):
    """
    Given a list of movie ids, return a list containing the ratings of those movies.
    """
    ratings = []
    
    for datum in data:
        if int(datum[1]) in movie_ids:
            ratings.append(datum[2])
            
    return ratings
```

```python
num_ratings = np.array([0] * 1683)
num_ratings[0] = 1

for i in range(len(data)):
    num_ratings[(int) (data[i, 1])] += 1
    
most_pop = num_ratings.argsort()[-10:][::-1]
plt.hist(get_ratings(most_pop), bins=5, range=(1,6))
plt.title('Ratings of 10 Most Popular')
plt.show()
```

```python
total_rating = np.array([0] * 1683)

for i in range(len(data)):
    total_rating[(int) (data[i, 1])] += (int) (data[i, 2])

avg_rating = total_rating / num_ratings
    
highest_rated = avg_rating.argsort()[-10:][::-1]
plt.hist(get_ratings(highest_rated), bins=5, range=(1,6))
plt.title('Ratings of 10 Highest Rated')
plt.show()
```

```python
import codecs
movies = [[]]
f = open('data/movies.txt', 'r', encoding='latin1')

for line in f:
    movies.append([int(x) for x in line.strip().split('\t')[2:]])
    
movies = np.array(movies)
```

```python
genre_names = ['unknown', 'Action', 'Adventure', 'Animation']

def get_ratings_for_genre(genre):
    movies_of_genre = []
    
    for i in range(1, len(movies)):
        if movies[i][genre]:
            movies_of_genre.append(i)
    
    plt.hist(get_ratings(movies_of_genre), bins=5, range=(1,6))
    plt.title('Ratings for %s Genre' % genre_names[genre])
    plt.show()
    
get_ratings_for_genre(1)
get_ratings_for_genre(2)
get_ratings_for_genre(3)
```
